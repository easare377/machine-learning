```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
# keras imports for the dataset and building our neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, LSTM, RepeatVector, TimeDistributed
from keras.utils import np_utils
from matplotlib import pyplot as plt
from attention_decoder import AttentionDecoder
```

```python
def generate_sequenceX():
    rand = np.random.randint(0, 1000)
    x = [0]*5
    y = [0]*5
    for i in range(len(x)):
        x[i] = np.random.randint(0, 1000)
        if i == 0:
            y[i] = x[i] * 2
        elif i == 1:
            y[i] = x[i] - 1
        elif i == 2:
            y[i] = (x[i] / 2) +  1
        else:
            y[i] = 0
    return x, y

def generate_all_sequence(count):
    X = [0]*count
    Y = [0] * len(X)
    for i in range(len(X)):
        X[i], Y[i] = generate_sequenceX()
    return X, Y
    
```

```python
model = Sequential()

# encoder layer
model.add(LSTM(100, activation='relu', input_shape=(3, 1)))

# repeat vector
model.add(RepeatVector(3))

# decoder layer
model.add(LSTM(100, activation='relu', return_sequences=True))

model.add(TimeDistributed(Dense(1)))
model.compile(optimizer='adam', loss='mse')

print(model.summary())


```

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3)
```
