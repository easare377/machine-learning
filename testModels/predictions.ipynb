```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
# keras imports for the dataset and building our neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, LSTM
from keras.utils import np_utils
from matplotlib import pyplot as plt
```

```python
x = list()
y = list()
x = [x+3 for x in range(-2, 10000, 3)]

for i in x:
    output_vector = list()
    output_vector.append(i+1)
    output_vector.append(i+2)
    y.append(output_vector)
    
print(x)
print(y)
```

```python
x = np.array(x).reshape(3334, 1, 1)
y = np.array(y)
print(x)
```

```python

# building a linear stack of layers with the sequential model
model = Sequential()
# hidden layer
model.add(LSTM(50, activation='relu', input_shape=(1,1)))
model.add(Dense(50, input_shape=(1,), activation='relu'))
# output layer
model.add(Dense(2))
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x, y, epochs=100, validation_split=0.2, batch_size=3)
```

```python
x_test = np.array([100000]).reshape(1, 1, 1)
x_test
```

```python

pred = model.predict(x_test)
np.around(pred[0])
```
