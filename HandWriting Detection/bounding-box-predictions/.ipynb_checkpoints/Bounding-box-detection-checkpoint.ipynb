```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten
from keras.utils import np_utils
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
```

```python
x = [1, 4, 7, 100, 8, 5, 0, 45]
y = [99, 96, 93, 0, 92, 100, 0, 55]
```

```python
x = np.array(x).reshape(8, 1, 1)
y = np.array(y).reshape(8, 1, 1)
x
```

```python
model = Sequential()

model.add(Dense(8, input_shape=(1, 1)))
model.add(Dense(1))
print(model.summary())
model.compile(optimizer='adam', loss='mse')

# wandb.init(project='simple-lstm')

```

```python
model.fit(x, y, epochs=5000, batch_size=3)
```

```python
test_output = model.predict(np.array([30]).reshape(1, 1, 1), verbose=0)
print(test_output)
```
